```python
def split_breastcancer_data(df):
    '''
    This function performs split on breast cancer data, stratify status.
    Returns train, validate, and test dfs.
    '''
    train_validate, test = train_test_split(df, test_size=.2, 
                                        random_state=123, 
                                        stratify=df.Status)
    train, validate = train_test_split(train_validate, test_size=.3, 
                                   random_state=123, 
                                   stratify=train_validate.Status)
    return train, validate, test
```

```python
train, validate, test = split_breastcancer_data(breast_cancer)
```

```python
pd.crosstab(train.Race, train.Status, normalize="index").plot.bar()
```

    -Black patients, despite being the smallest in number of patients, have the highest number of deaths 
    -Other race has the most patients and also the fewest deaths 

```python
pd.crosstab(train['Marital Status'], train.Status, normalize="index").plot.bar()
```
